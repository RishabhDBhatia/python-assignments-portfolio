```python
# defining variables (Graders note: please take the commenting below to be consistent in repeated cells below- thanks! )
x0 = 0.1
y0 = 0.1
alpha = 0.1
num_iterations = 10

# defining the function
def gradient_descent (x0, yo, fun_1, grad_fun_1, alpha, num_iterations): #consider all variable inputs
    x = x0 # initialize
    y = y0 # initialize

# making a for loop for
    for i in range(num_iterations):
        gradient_x, gradient_y = grad_fun_1 (x, y)
        
        x = x - alpha * gradient_x
        y = y - alpha * gradient_y

    return x, y

#the function as asked in kritik 8 
def fun_1(x, y) :
    return x**2 + y**2
    
#partial derivatives (visually determined)
def grad_f_1(x, y) :
    grad_x = 2*x
    grad_y = 2*y
    
    return grad_x, grad_y

result = gradient_descent (x0, y0, fun_1, grad_f_1, alpha, num_iterations) # calling the function
print ("Optimal values:", result) # print the optimal value
```
