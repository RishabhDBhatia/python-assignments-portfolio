```python
import sympy as sp

x, y = sp.symbols('x y')

# Define a function
f = sp.exp(x)*sp.sin(y) + y**3

# find partial derivative wrt x
df_dx = sp.diff(f, x)
print("df/dx =", df_dx)

# find partial derivative wrt y
df_dy = sp.diff(f, y)
print("df/dy =", df_dy)
```

```python
import sympy as sp

x, y = sp.symbols('x y')  # define x and y in sympy

# Define a function
f = (x**2)*y + x*(y**2)

# Calculate partial derivatives
df_dx = sp.diff(f, x)
df_dy = sp.diff(f, y)
gradient = df_dx + df_dy

# Define point
x_val = 1
y_val = -1

def gradient_at_point(x, y):  # defining gradient at point
    return x**2 + 2*x*y

print("Gradient at (1, -1) is:", gradient_at_point(x_val, y_val), gradient_at_point(x_val, y_val))

def magnitude(x, y):  # define magnitude
    return sp.sqrt((gradient_at_point(x, y))**2 + (gradient_at_point(x, y))**2)

print("Magnitude at (1, -1) is:", magnitude(x_val, y_val))
```

```python
import sympy as sp

# Define symbols
x, y = sp.symbols('x y')

# Define the function
f = sp.log(x**2 + y**2)

# Calculate second partial derivatives
second_derivative_x = sp.diff(sp.diff(f, x), x)  # Second partial derivative with respect to x
second_derivative_y = sp.diff(sp.diff(f, y), y)  # Second partial derivative with respect to y
mixed_derivative = sp.diff(sp.diff(f, x), y)     # Mixed partial derivative

# Output the results
print("Second derivative with respect to x:", second_derivative_x)
print("Second derivative with respect to y:", second_derivative_y)
print("Mixed partial derivative:", mixed_derivative)

# Comment on symmetry
print("\nAccording to Clairaut's Theorem, the mixed partial derivatives (fxy and fyx) are symmetrical.")
print("Hence, fxy(x, y) =", mixed_derivative, "and fyx(x, y) =", mixed_derivative, "will be the same.")
```

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Define symbols
x, y = sp.symbols('x y')

# Define the function
j = x**2 - 2*x*y + y**2

# Convert the function to a numpy-compatible function
j_func = sp.lambdify((x, y), j, 'numpy')

# Generate x and y values
x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)

# Create a meshgrid
X, Y = np.meshgrid(x_vals, y_vals)

# Evaluate the function on the meshgrid
Z = j_func(X, Y)

# Create the contour plot
plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of j(x, y) = x^2 - 2xy + y^2')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

```
